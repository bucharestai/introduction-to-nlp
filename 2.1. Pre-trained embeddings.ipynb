```
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
```

```
elmo = hub.Module("https://tfhub.dev/google/elmo/2")
```

```
messages = [
    "i went to the bank to withdraw some cash", 
    "i went to the bank of the river", 
    "my neighbour works as a teller in a bank"]
```

```
with tf.Session() as session:
  session.run([tf.global_variables_initializer(), tf.tables_initializer()])

  outputs = session.run(elmo(
    messages,
    signature="default",
    as_dict=True))
  
  word_embeddings = outputs["elmo"] 
  
  for i, message_embedding in enumerate(np.array(word_embeddings).tolist()):
    print("Message: {}".format(messages[i]))
    print("Embedding size: {}".format(len(message_embedding)))
    for e in message_embedding:
      print(e)
```
